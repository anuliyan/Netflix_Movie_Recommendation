```python
plt.figure(figsize=(10,6))
sns.boxplot(x='user rating score', data=Dataframe)
plt.title('Released Year')
```

```python
plt.figure(figsize=(20,6))
sns.barplot(x='user rating score',y='ratingDescription', data=Dataframe)
plt.title('Released Year')
```

```python
# Correlation Matrix

# Compute the correlation matrix
d= Dataframe[['ratingDescription','release year','user rating score']]

corr = d.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, 
            square=True, linewidths=.5, annot=True, cmap=cmap)
plt.yticks(rotation=0)
plt.title('Correlation Matrix of all Numerical Variables')
plt.show()
```

# User Rating Score by category and year

```python
plt.subplots(figsize=(12,10))
max_ratings=Dataframe.groupby('rating')['rating'].count()
max_ratings=max_ratings[max_ratings.values>50]
max_ratings.sort_values(ascending=True,inplace=True)
mean_shows=Dataframe[Dataframe['rating'].isin(max_ratings.index)]
piv=mean_shows.groupby(['release year','rating'])['user rating score'].mean().reset_index()
piv=piv.pivot('release year','rating','user rating score')
sns.heatmap(piv,annot=True,cmap='YlGnBu')
plt.title('Average User Score By Rating')
plt.show()
```
